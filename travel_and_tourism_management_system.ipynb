```python
# ==============================
# TRAVEL & TOURISM MANAGEMENT SYSTEM
# ==============================

import sqlite3
from datetime import datetime

# Connect to SQLite Database
conn = sqlite3.connect("travel_tourism.db")
cursor = conn.cursor()

# ==============================
# DATABASE TABLES
# ==============================

cursor.execute("""
CREATE TABLE IF NOT EXISTS users(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    email TEXT UNIQUE,
    password TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS packages(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    package_name TEXT,
    destination TEXT,
    price INTEGER,
    duration TEXT
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS bookings(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    user_email TEXT,
    package_name TEXT,
    booking_date TEXT
)
""")

conn.commit()

# ==============================
# FUNCTIONS
# ==============================

def register():
    name = input("Enter Name: ")
    email = input("Enter Email: ")
    password = input("Enter Password: ")

    try:
        cursor.execute("INSERT INTO users (name,email,password) VALUES (?,?,?)",
                       (name,email,password))
        conn.commit()
        print("✅ Registration Successful!\n")
    except:
        print("❌ Email already exists!\n")

def login():
    email = input("Enter Email: ")
    password = input("Enter Password: ")

    cursor.execute("SELECT * FROM users WHERE email=? AND password=?",
                   (email,password))
    user = cursor.fetchone()

    if user:
        print("✅ Login Successful!\n")
        user_menu(email)
    else:
        print("❌ Invalid Credentials!\n")

def add_package():
    name = input("Package Name: ")
    destination = input("Destination: ")
    price = int(input("Price: "))
    duration = input("Duration (e.g. 5 Days): ")

    cursor.execute("INSERT INTO packages (package_name,destination,price,duration) VALUES (?,?,?,?)",
                   (name,destination,price,duration))
    conn.commit()
    print("✅ Package Added Successfully!\n")

def view_packages():
    cursor.execute("SELECT * FROM packages")
    packages = cursor.fetchall()

    print("\n📦 Available Packages:\n")
    for pkg in packages:
        print(f"ID: {pkg[0]} | {pkg[1]} | {pkg[2]} | ₹{pkg[3]} | {pkg[4]}")
    print()

def book_package(email):
    view_packages()
    package_id = int(input("Enter Package ID to Book: "))

    cursor.execute("SELECT package_name FROM packages WHERE id=?",
                   (package_id,))
    package = cursor.fetchone()

    if package:
        booking_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        cursor.execute("INSERT INTO bookings (user_email,package_name,booking_date) VALUES (?,?,?)",
                       (email,package[0],booking_date))
        conn.commit()
        print("✅ Booking Confirmed!\n")
    else:
        print("❌ Invalid Package ID!\n")

def view_bookings():
    cursor.execute("SELECT * FROM bookings")
    bookings = cursor.fetchall()

    print("\n📑 All Bookings:\n")
    for b in bookings:
        print(f"User: {b[1]} | Package: {b[2]} | Date: {b[3]}")
    print()

def user_menu(email):
    while True:
        print("1. View Packages")
        print("2. Book Package")
        print("3. Logout")

        choice = input("Choose option: ")

        if choice == "1":
            view_packages()
        elif choice == "2":
            book_package(email)
        elif choice == "3":
            break
        else:
            print("Invalid choice!")

def admin_menu():
    while True:
        print("1. Add Package")
        print("2. View Packages")
        print("3. View Bookings")
        print("4. Logout")

        choice = input("Choose option: ")

        if choice == "1":
            add_package()
        elif choice == "2":
            view_packages()
        elif choice == "3":
            view_bookings()
        elif choice == "4":
            break
        else:
            print("Invalid choice!")

# ==============================
# MAIN MENU
# ==============================

while True:
    print("===== TRAVEL & TOURISM MANAGEMENT =====")
    print("1. Register")
    print("2. Login")
    print("3. Admin Login")
    print("4. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        register()
    elif choice == "2":
        login()
    elif choice == "3":
        password = input("Enter Admin Password: ")
        if password == "admin123":
            admin_menu()
        else:
            print("❌ Wrong Admin Password!\n")
    elif choice == "4":
        print("Thank you for using the system!")
        break
    else:
        print("Invalid choice!\n")
```
