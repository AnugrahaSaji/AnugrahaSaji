```python
# ==========================================
# DIGITAL COLLEGE BUS TRANSPORT SYSTEM
# ==========================================

import sqlite3
import random
import time
from datetime import datetime

# Connect Database
conn = sqlite3.connect("college_bus_system.db")
cursor = conn.cursor()

# ==========================================
# CREATE TABLES
# ==========================================

cursor.execute("""
CREATE TABLE IF NOT EXISTS students(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    reg_no TEXT UNIQUE,
    rfid TEXT UNIQUE
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS bus_logs(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    rfid TEXT,
    student_name TEXT,
    bus_no TEXT,
    location TEXT,
    time TEXT
)
""")

conn.commit()

# ==========================================
# GPS SIMULATION FUNCTION
# ==========================================

def get_bus_location():
    locations = [
        "College Gate",
        "Main Road",
        "City Junction",
        "Bus Stop 1",
        "Bus Stop 2",
        "Highway",
        "Campus Entrance"
    ]
    return random.choice(locations)

# ==========================================
# FUNCTIONS
# ==========================================

def register_student():
    name = input("Enter Student Name: ")
    reg_no = input("Enter Register Number: ")
    rfid = input("Enter RFID Card Number: ")

    try:
        cursor.execute("INSERT INTO students (name,reg_no,rfid) VALUES (?,?,?)",
                       (name,reg_no,rfid))
        conn.commit()
        print("✅ Student Registered Successfully\n")
    except:
        print("❌ RFID or Register Number Already Exists\n")

def scan_rfid():
    rfid = input("Scan RFID Card: ")

    cursor.execute("SELECT name FROM students WHERE rfid=?", (rfid,))
    student = cursor.fetchone()

    if student:
        location = get_bus_location()
        time_now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        cursor.execute("""
        INSERT INTO bus_logs (rfid,student_name,bus_no,location,time)
        VALUES (?,?,?,?,?)
        """, (rfid,student[0],"Bus-1",location,time_now))

        conn.commit()

        print(f"🚌 Welcome {student[0]}")
        print(f"📍 Current Bus Location: {location}")
        print(f"⏰ Time: {time_now}\n")
    else:
        print("❌ Invalid RFID Card\n")

def view_students():
    cursor.execute("SELECT * FROM students")
    students = cursor.fetchall()

    print("\n📘 Registered Students\n")
    for s in students:
        print(f"Name: {s[1]} | Reg No: {s[2]} | RFID: {s[3]}")
    print()

def view_logs():
    cursor.execute("SELECT * FROM bus_logs")
    logs = cursor.fetchall()

    print("\n📊 Travel Logs\n")
    for log in logs:
        print(f"Student: {log[2]} | Bus: {log[3]} | Location: {log[4]} | Time: {log[5]}")
    print()

# ==========================================
# MAIN MENU
# ==========================================

while True:
    print("====== DIGITAL COLLEGE BUS SYSTEM ======")
    print("1. Register Student (Admin)")
    print("2. Scan RFID Card (Bus Entry)")
    print("3. View Students")
    print("4. View Travel Logs")
    print("5. Exit")

    choice = input("Select Option: ")

    if choice == "1":
        register_student()
    elif choice == "2":
        scan_rfid()
    elif choice == "3":
        view_students()
    elif choice == "4":
        view_logs()
    elif choice == "5":
        print("System Closed")
        break
    else:
        print("Invalid Option\n")
```
